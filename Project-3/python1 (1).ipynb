```python
# we create a function to check if the data is proportional based on a pre-defined variable.
def counts(df, var):
    c = df[var].value_counts(dropna=False)
    p = df[var].value_counts(dropna=False, normalize=True)
    return pd.concat([c,p], axis=1, keys=['counts', '%'])
```

```python
counts(df_sample3,"water_insecurity_index")
```

```python
counts(df_sample3,"district")
```

Cluster Sampling

```python
df_sample4 = pd.read_csv("/content/kathmandu_household.csv")
df_sample4.head(5)
```

```python
df_sample4.info()
```

```python
length = len(df_sample4)
groups = 10
elements = length / groups
print('elements by group:',elements)
```

```python
group_list = []
group_id = 0
element_count = 0

for _ in df.iterrows():
    group_list.append(group_id)
    element_count += 1
    if element_count > elements:
        element_count = 0
        group_id += 1

#print(group_list) to see all group list generated.
np.unique(group_list, return_counts=True)
```

```python
np.shape(group_list), df.shape
```

```python
df['group'] = group_list
display(df.head(3))
```

```python
import random as rd
rd.seed(None)
group_selected = rd.randint(0, groups - 1)
print('group selected:',group_selected)
```

```python
df_cluster_sample = df[df['group'] == group_selected]

df_cluster_sample.shape
```
