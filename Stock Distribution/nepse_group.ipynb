```python
import pandas as pd

df = pd.read_csv("/content/subset.csv")
new_df = df.head(10)
print(new_df.tail(5))
```

1. Range

```python
volume = new_df['volume'].to_list() # dataframe lai list ma convert garim
max_vol = new_df['volume'].max() #max value
min_vol = new_df['volume'].min() #min value
range = max_vol-min_vol
print(f"Range: {range}") # range
```

```python
# variance = new_df['volume'].var()
# print(variance)
```

2. Variance

```python
mean = new_df['volume'].sum()/len(volume)
print(f"Mean: {mean}")
sum =0
for x in volume:
  sum += (x-mean)**2
  # print(sum)
print(f"Sum: {sum}")
#variance = sigma(x-mean)^2/(N-1) for sample variance
variance = sum/(len(volume) - 1)
print(f"Sample Variance: {variance}")
```

3. Standard Deviation

```python
sd = variance**0.5
print(f"Standard deviation: {sd}")
```

```python
z = (3507.0 - mean)/ sd
print(f"Z-score: {z}")
```
